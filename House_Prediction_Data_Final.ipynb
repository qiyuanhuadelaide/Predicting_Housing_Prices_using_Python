```python
# put data into two categories, one is train
# and the other is test based on if there's any missing value in the SalePrice column
# delete the SalePrice column 
train=df[df['SalePrice'].notnull()]
test=df[df['SalePrice'].isnull()]
test=test.drop(['SalePrice'],axis=1)
train.shape,test.shape
```

```python
def get_missing_data(df):
    total = df.isnull().sum()[df.isnull().sum()!=0].sort_values(ascending=False)
    percent = (total/df.isnull().count()[total.index]).sort_values(ascending=False)
    missing_data = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
    return missing_data

# calculate the missing values in the train and order it descendingly  
train_missing_data=get_missing_data(train)
train_missing_data
```

```python
# calculate the missing values in the test and order it descendingly
test_missing_data = get_missing_data(test)
test_missing_data
```

```python
# some columns, such as Alley,FireplaceQu,PoolQC,Fence,MiscFeature, has many missing values which should be deleted.
# column id is useless so delete it as well.
test_Id=test[['Id']]
train=train.drop(['Id','Alley','FireplaceQu','PoolQC','Fence','MiscFeature'],axis=1)
test=test.drop(['Id','Alley','FireplaceQu','PoolQC','Fence','MiscFeature'],axis=1)
# show the shape attribute of data(number of rows, number of columns) 
train.shape,test.shape
```

```python
import numpy as np

def fill_missing_data(df):
    # all the missing columns 
    missing_columns=df.isnull().sum()[df.isnull().sum()!=0].index
    # all the missing columns among metric data
    missing_columns_numeric=df.select_dtypes(include=[np.number]).columns
    # all the missing columns among non metric data
    missing_columns_category=df.select_dtypes(exclude=[np.number]).columns
    # fill up the missing values
    for column in missing_columns_numeric:
        df.loc[:,column]=df[column].fillna(df[column].mean())
    for column in missing_columns_category:
        df.loc[:,column]=df[column].fillna(df[column].mode()[0])
    
    return df
train=fill_missing_data(train)
train.isnull().sum().sum()
```

```python
test=fill_missing_data(test)
test.isnull().sum().sum()
```

```python
# code catergory label 
from sklearn.preprocessing import LabelEncoder

labelencode=LabelEncoder()
columns_category=train.select_dtypes(exclude=[np.number]).columns

for column in columns_category:
    train[column]=labelencode.fit_transform(train[column])
    test[column]=labelencode.fit_transform(test[column])
```

```python
# data standardization since the dimension of data is inconsistent
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
train.iloc[:,:-1] = stdsc.fit_transform(train.iloc[:,:-1])
test = stdsc.transform(test)
```

```python
# Splitting data into training and testing sets
# cannot being used for evaluating the effectiveness of the model since test does not have label
# so need to divide training data set 
from sklearn.model_selection import train_test_split

X, y = train.iloc[:, :-1], train.iloc[:, -1]
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# linear regression, and visualize predicted value and error
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

lr = LinearRegression()
lr.fit(X_train, y_train)# train
# predict results on the test dataset 
y_pred = lr.predict(X_test)

plt.scatter(y_test,  y_pred - y_test,
            c='limegreen', marker='s', edgecolor='white',
            label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MSE : %.3f, R^2 : %.3f' % (mse,r2))
```

```python
# statistically analyzed house price and all the features with Pearson correlation 
def spearman(df,columns):
    spr = pd.DataFrame()
    spr['featrue'] = columns
    spr['corr'] = [df[f].corr(df['SalePrice'], 'spearman') for f in columns]
    spr = spr.sort_values('corr')
    plt.figure(figsize=(6, 0.3*len(columns)))
    sns.barplot(data= spr, y='featrue', x='corr', orient='h')
    
columns_numeric = train.select_dtypes(include=[np.number]).columns    
spearman(train,columns_numeric)
```
